```python
def get_result(desc):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": desc},
        ])
    return response
```

```python
def get_json(res):
    return json.loads(res['choices'][0]['message']['content'])
```

```python
import time
from tqdm import tqdm

df['product_type']=None
df['colors'] = None
df['brand'] = None
df['hardware'] = None
df['material'] = None
df['year'] = None


for i, row in tqdm(df.iterrows(), total=df.shape[0]):
    while True:
        try:
            res = get_result(df.iloc[i]['description'])
            data = get_json(res)
            df.at[i,'product_type'] = data['product_type']
            df.at[i,'colors'] = data['colors']
            df.at[i,'brand'] = data['brand']
            df.at[i,'hardware'] = data['hardware']
            df.at[i,'material'] = data['material']
            df.at[i,'year'] = data['year']
            break
        except:
            time.sleep(1)
df.head()
```
